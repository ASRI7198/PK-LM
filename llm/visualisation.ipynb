```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
file_path = "message.txt"
with open(file_path, 'r') as file:
    data_text = file.read()
```

```python
pattern = re.compile(r"'loss': ([\d.]+), 'grad_norm': ([\d.]+), 'learning_rate': [\d.e-]+, 'epoch': ([\d.]+)")
matches = pattern.findall(data_text)
```

```python
data = pd.DataFrame(matches, columns=['loss', 'grad_norm', 'epoch'])
data = data.astype(float)  
```

```python
plt.figure(figsize=(10, 6))

plt.plot(data['epoch'], data['loss'], label='Loss', marker='o', color='blue')

plt.title('Training Progress (Loss)')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.ylim(data['loss'].min() - 0.1, data['loss'].max() + 0.1)  # Zoom sur les valeurs de la perte
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(data['epoch'], data['grad_norm'], label='Gradient Norm', marker='o', color='orange')

plt.title('Training Progress (Gradient Norm)')
plt.xlabel('Epoch')
plt.ylabel('Gradient Norm')
plt.ylim(data['grad_norm'].min() - 0.1, data['grad_norm'].max() + 0.1)  # Zoom sur les valeurs de la norme des gradients
plt.legend()
plt.grid(True)
plt.show()
```
